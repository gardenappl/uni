**Generating data**

```python
import random
```

```python
def make_row():
  # math, physics, Ukrainian
  exams = [
           random.randrange(5) + 1,
           random.randrange(5) + 1, 
           random.randrange(5) + 1
          ]
  privilege = random.choice([0,1])
  if privilege == 0:
    accepted = sum(exams) >= 11 and exams[0] > 3 and exams[1] > 3
  else:
    accepted = exams[0] >= 3 and exams[1] >= 3 and exams[2] >= 3
  return ([privilege] + exams, accepted)

input_train = []
output_train = []
input_val = []
output_val = []

print("Training set")
print("Privilege, math, phys, Ukr")
for i in range(1000):
  row = make_row()
  print(row)
  input_train.append(row[0])
  output_train.append(row[1])

```

```python
print("Validation set")
print("Privilege, math, phys, Ukr")
for i in range(200):
  row = make_row()
  print(row)
  input_val.append(row[0])
  output_val.append(row[1])
```

```python
import keras.models as M
import keras.layers as L
import keras.backend as K
import tensorflow as tf
import numpy as np
```

```python
input_train = np.array(input_train)
output_train = tf.keras.utils.to_categorical(np.array(output_train))

input_val = np.array(input_val)
output_val = tf.keras.utils.to_categorical(np.array(output_val))
```

```python
for layers in [1, 2, 3, 4, 5]:
  print("Layers: ", layers)

  K.clear_session()
  model = M.Sequential()
  model.add(L.Dense(4, activation='relu'))
  for i in range(layers):
    print("Adding layer ", i)
    model.add(L.Dense(128, activation='relu'))
  model.add(L.Dense(2, activation='sigmoid'))

  model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
  )

  model.fit(
    input_train,
    output_train,
    batch_size=5,
    epochs=50,
    validation_data=(input_val, output_val)
  )
```

Бачимо, що одного шару зі 128 нейронами достатньо, аби отримати задовільний реузльтат (99% точність на валідаційній вибірці), за умови, що задана достатня кількість епох.

Спробуємо зменшити кількість нейронів, але збільшити кількість епох.

```python
for neurons in [20, 40, 60, 80, 100, 120]:
  print("Neurons in hidden layer: ", neurons)

  K.clear_session()
  model = M.Sequential()
  model.add(L.Dense(4, activation='relu'))
  model.add(L.Dense(neurons, activation='relu'))
  model.add(L.Dense(2, activation='sigmoid'))

  model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
  )

  model.fit(
    input_train,
    output_train,
    batch_size=5,
    epochs=150,
    validation_data=(input_val, output_val)
  )
```

Знову бачимо `val_accuracy = 0.99`, а у випадку мережі зі 40 нейронами бачимо `val_accuracy = 1.0`. Це свідчить про те, що кількість нейронів (за умови, шо їх більше 20) на результат впливає менше, ніж кількість епох.
Спробуємо знову використати 20 нейронів, ще раз збільшивши кількість епох.

```python
K.clear_session()
model = M.Sequential()
model.add(L.Dense(4, activation='relu'))
model.add(L.Dense(20, activation='relu'))
model.add(L.Dense(2, activation='sigmoid'))

model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)

model.fit(
  input_train,
  output_train,
  batch_size=1,
  epochs=200,
  validation_data=(input_val, output_val)
)
```

Бачимо, що точність як і для навчальної, так і для тестової вибірки коливається між 99% та 100%.

Можливо, було би реалістично отримати точність 100%, оскільки у нас обмежена кількість комбінацій вхідних даних (якщо припустити п'ятибальну шкалу оцінювання: `5 * 5 * 5 * 2 == 250`). Для 250 комбінацій напевно можна було би виконати overfitting, і саме таким чином отримати абсолютну точність.

Проте ми використовуємо **мінімальну кількість шарів та нейронів**, що запобігає overfitting (і відповідає умові лабораторної), тому маємо точність ~99%, яку будемо вважати задовільною.
